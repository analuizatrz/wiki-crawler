# Convertion of wikihtml to html | Documentation
### After collecting the content of a page, convert the wikihtml file to html

The algorithm makes a bridge between a Java Class that converts the content of the file and a python algorithm wich sends the content to the gateway and stores it.

# Imported modules

```python
from py4j.java_gateway import JavaGateway
import os
from wkcollect import collect_all
from wkio import write_file, create_folder_if_does_not_exist, create_file_if_does_not_exist, append_file
from wkutils import create_logger, Params
import pandas as pd
```

# Create the Java Server

Execute EntryPoint.java in /htmlConverterEntryPoint and the server will be started

# Connect to the Java Server

```python
gateway = JavaGateway()

converter = gateway.entry_point.getConverter()
gateway.jvm.java.lang.System.out.println('Conectado !')
```

# Convertion

convert_content() extrats the content of the file and calls the function of the Java Class that returns the converted content and then write the file.

```python
def convert_content(title, params, folder_data, log):
    input = open(f"{params.input_folder}/{title}").read()
    html = converter.convert(input)
    write_file(f"{folder_data}/{title}.html", html)

def convert_all(input_folder, output_folder, params):
    create_folder_if_does_not_exist(output_folder)
    params.input_folder = input_folder
    titles = os.listdir(input_folder)
    collect_all(titles, convert_content, params, output_folder, create_logger(output_folder))

base_folder = "/home/sinval/Documentos/Wiki-Coletor/Wiki_Colector_Notebooks/Collector/"
output_folder = f"{base_folder}/converted_html_GA"
input_folder = f"{base_folder}/collected_pages_GA/data"
titles = os.listdir(input_folder)
convert_all(input_folder, output_folder, Params())
```
